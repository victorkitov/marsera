# Эксперименты

```python
import numpy as np
import matplotlib.pyplot as plt
from mars_module import Earth
```

```python
def plot(X, y, y_pred, max_terms, ind=1):
    plt.figure()
    plt.plot(X[:, ind], y, 'r.', label=f'train with dim {X.shape}')
    plt.plot(X[:, ind], y_pred, 'b.', label='predict')
    plt.xlabel(f'$x_{ind}$')
    plt.ylabel('$y$')
    plt.title(f'$max\_terms={max_terms}$')
    plt.legend()
    plt.show()


def fit(X, y, max_terms=1, ind=1):
    model = Earth(max_terms=max_terms)
    model.forward_pass(X, y)

    print(model.term_list)
    
    y_pred = model.predict(X)

    plot(X, y, y_pred, max_terms=max_terms, ind=ind)
```

### Галочки и их суммы

#### Галочка

```python
np.random.seed(0)
m = 100
n = 3
X = 80 * np.random.uniform(size=(m, n)) - 40
y = np.abs(X[:, 1] - 4.0) + 1 * np.random.normal(size=m)

plt.figure()
plt.plot(X[:, 1],y, 'r.')
plt.xlabel('x_1')
plt.ylabel('y')
plt.title('100 сэмплов, 3 признака, галочка по 1-ой координате (здесь и далее с нуля), остальные 2 - шум')
plt.show()
```

```python
fit(X, y, max_terms=1, ind=1)
#0.2s
```

```python
fit(X, y, max_terms=3, ind=1)
#0.2s
```

```python
fit(X, y, max_terms=5, ind=1)
#0.3s
```

#### Сумма 3х галочек

```python
np.random.seed(0)
m = 1000
n = 10
X = 80 * np.random.uniform(size=(m, n)) - 40
y = 0
t = np.random.randint(-10, high=10, size=5)
for i in range(5):
    y += np.abs(X[:, i] - t[i]) + 1 * np.random.normal(size=m)

plt.figure()
plt.plot(X[:, 1], y, 'r.')
plt.xlabel('$x_1$')
plt.ylabel('$y$')
plt.title('1000 сэмплов, 10 признаков, сумма галок (с разными порогами) по первым 5 координатам с дисперсией 1, остальные 5 - шум')
plt.show()
```

```python
fit(X, y, max_terms=1, ind=1)
# 0.2s
```

```python
fit(X, y, max_terms=3, ind=1)
# 24.8s
```

```python
fit(X, y, max_terms=13, ind=1)
#8m42.0s
```
