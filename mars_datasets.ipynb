### Галочки и их суммы

```python
np.random.seed(0)
m = 1000
n = 10
X = 80*np.random.uniform(size=(m,n)) - 40
y = np.abs(X[:,6] - 4.0) + 1*np.random.normal(size=m)

plt.figure()
plt.plot(X[:,6],y,'r.')
plt.xlabel('x_6')
plt.ylabel('y')
plt.title('10000 сэмплов, 10 признаков, галочка по 6-ой координате (здесь и далее с нуля), остальные 9 - шум (пример из py-earth)')
plt.show()
```

```python
np.random.seed(0)
m = 10000
n = 10
X = 80*np.random.uniform(size=(m,n)) - 40
y = np.abs(X[:,6] - 4.0) + np.abs(X[:,7] + 20.0) + np.abs(X[:,8] + 40.0) + 1*np.random.normal(size=m)

plt.figure()
plt.plot(X[:,6],y,'r.')
plt.xlabel('x_6')
plt.ylabel('y')
plt.title('10000 сэмплов, 10 признаков, сумма трёх галочек с разными сдвигами по 6, 7 и 8-ой координате, остальные 7 - шум')
plt.show()
```

```python
np.random.seed(0)
m = 1000
n = 5
X = 200*np.random.uniform(size=(m,n)) - 100
y = np.abs(X[:,1] - 50.0) + 8*np.random.normal(size=m)

plt.figure()
plt.plot(X[:,1],y,'r.')
plt.xlabel('x_1')
plt.ylabel('y')
plt.title('1000 сэмплов, 5 признаков, галка по 1-ой координате с большой дисперсией, остальные 4 - шум')
plt.show()
```

```python
np.random.seed(0)
m = 100000
n = 100
X = 2000*np.random.uniform(size=(m,n)) - 1000
y = 0
t = np.random.randint(-1000, high=1000, size=50)
for i in range(50):
    y += np.abs(X[:,i] - t[i]) + 1*np.random.normal(size=m)

plt.figure()
plt.plot(X[:,1],y,'r.')
plt.xlabel('x_1')
plt.ylabel('y')
plt.title('100000 сэмплов, 100 признаков, сумма галок (с разными порогами) по первым 50 координатам с дисперсией 1, остальные 50 - шум')
plt.show()
```

### Другие функции

```python
np.random.seed(0)
m = 1000
n = 10
X = 10*np.random.uniform(size=(m,n)) - 5
y = np.sin(X[:,6]) + 0.1*np.random.normal(size=m)

plt.figure()
plt.plot(X[:,6],y,'r.')
plt.xlabel('x_6')
plt.ylabel('y')
plt.title('1000 сэмплов, 10 признаков, синус по 6-ой координате, остальные 9 - шум (пример из py-earth)')
plt.show()
```

```python
np.random.seed(0)
m = 1000
n = 10
X = 10*np.random.uniform(size=(m,n)) - 5
y = np.sin(X[:,6]) + np.abs(X[:,7] - 1) + 0.01*np.random.normal(size=m)

plt.figure()
plt.plot(X[:,6],y,'r.')
plt.xlabel('x_6')
plt.ylabel('y')
plt.title('1000 сэмплов, 10 признаков, синус по 6-ой координате, галочка по 7-ой, остальные 8 - шум (пример из py-earth)')
plt.show()
```

```python
np.random.seed(0)
m = 1000
n = 10
X = 10*np.random.uniform(size=(m,n)) - 5
y = np.sin(X[:,6]) + np.exp(X[:,7]/10) + 0.01*np.random.normal(size=m)

plt.figure()
plt.plot(X[:,6],y,'r.')
plt.xlabel('x_6')
plt.ylabel('y')
plt.title('1000 сэмплов, 10 признаков, синус по 6-ой координате, маленькая экспонента по 7-ой, остальные 8 - шум (пример из py-earth)')
plt.show()
```

```python
np.random.seed(0)
m = 1000
n = 10
X = 10*np.random.uniform(size=(m,n)) - 5
y = np.abs(X[:,6]) + np.exp(X[:,7]/4) + 0.01*np.random.normal(size=m)

plt.figure()
plt.plot(X[:,6],y,'r.')
plt.xlabel('x_6')
plt.ylabel('y')
plt.title('1000 сэмплов, 10 признаков, галка по 6-ой координате, экспонента по 7-ой, остальные 8 - шум (пример из py-earth)')
plt.show()
```

```python
np.random.seed(0)
m = 1000
n = 10
X = 10*np.random.uniform(size=(m,n)) - 5
y = X[:,6] ** 2 + 0.1*np.random.normal(size=m)

plt.figure()
plt.plot(X[:,6],y,'r.')
plt.xlabel('x_6')
plt.ylabel('y')
plt.title('1000 сэмплов, 10 признаков, x^2 по 6-ой координате, остальные 9 - шум (пример из py-earth)')
plt.show()
```

```python
np.random.seed(0)
m = 10000
n = 10
X = 100*np.random.uniform(size=(m,n)) - 50
y = (X[:,6] - 5) ** 3 + 100*(X[:,7] + 3) ** 2 + 1000*(X[:,8] - 4) + 0.1*np.random.normal(size=m)

plt.figure()
plt.plot(X[:,6],y,'r.')
plt.xlabel('x_6')
plt.ylabel('y')
plt.title('10000 сэмплов, 10 признаков, (x - 5)^3 + 100(x + 3)^2 + 10000(x - 4) по 6, 7 и 8-ой координате соотв., остальные 7 - шум (пример из py-earth)')
plt.show()
```
