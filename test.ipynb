```python
import copy
import numpy as np
from mars_module import Earth
from matplotlib import pyplot as plt
from numpy import linalg as LA
from scipy.optimize import minimize, rosen
```

```python
#Create some fake data
np.random.seed(0)
m = 100
n = 3

X = 80 * np.random.uniform(size=(m, n)) - 40
y = np.abs(X[:, 0] - 4.0) + 1 * np.random.normal(size=m)

#Plot data
plt.figure()
plt.plot(X[:, 0],y,'r.')
plt.xlabel('x_1')
plt.ylabel('y')
plt.show()
```

```python
model = Earth(max_terms=3)
model.forward_pass(X, y)
```

```python
model.term_list
```

```python
model.coeffs
```

```python
y_hat = model.predict(X)
```

```python
#Plot the model
plt.figure()
plt.plot(X[:,0],y,'r.')
plt.plot(X[:,0],y_hat,'b.')
plt.xlabel('x_3')
plt.ylabel('y')
plt.title('Simple Earth Example')
plt.show()
```
